# Problem Definition
## Latar Belakang
Jamur merupakan organisme yang menguraikan bahan organik dan berperan dalam siklus nutrisi alami. Jamur dapat ditemukan di berbagai habitat, mulai dari tanah hingga air, dan parasit pada organisme lain. Beberapa jamur memiliki sifat edibel (dapat dimakan) dan digunakan sebagai bahan makanan, obat-obatan, atau bahan kimia, sementara yang lainnya beracun atau mematikan jika dikonsumsi oleh manusia atau hewan. Usulan untuk melakukan klasifikasi jamur yang dapat dimakan dan beracun didasarkan pada kepentingan akan keselamatan konsumen dan masyarakat umum. Maka dari itu, penting untuk memiliki sistem klasifikasi yang dapat membedakan jamur yang aman dan beracun secara cepat dan akurat. Selain itu, banyak masyarakat yang tertarik dalam mengumpulkan jamur untuk keperluan konsumsi atau hobi yang membuat pengetahuan tentang jenis jamur yang aman dan beracun semakin penting.

## Tujuan Penelitian
1. Meneliti cara mengidentifikasikan jamur yang dapat dimakan dan jamur beracun berdasarkan ciri-ciri morfologi dan karakteristiknya.
2. Membangun model klasifikasi yang memprediksi apakah suatu jamur dapat dimakan atau beracun berdasarkan karakteristik yang diamati.
3. Menentukan model klasifikasi terbaik untuk memprediksi apakah suatu jamur dapat dimakan atau beracun berdasarkan  evaluasi kinerja berbagai model.

## Rumusan Masalah
1. Bagaimana cara mengidentifikasi jamur yang dapat dimakan dan jamur beracun berdasarkan fitur-fitur morfologi dan karakteristik?
2. Bagaimana cara mengembangkan model klasifikasi menggunakan teknik machine learning untuk memprediksi apakah suatu jamur dapat dimakan atau beracun?
3. Bagaimana cara menentukan model klasifikasi yang paling optimal dalam memprediksi apakah suatu jamur dapat dimakan atau beracun?

## Metode
Algoritma untuk membangun model tugas klasifikasi, diantaranya
*   Decision Tree: Menggunakan aturan pemisahan berbasis fitur untuk membagi data ke dalam subset yang homogen.
*   Random Forest: Algoritma ensemble yang terdiri dari banyak decision trees, di mana hasil prediksi akhir diambil berdasarkan mayoritas suara atau rata-rata dari hasil prediksi individu dari setiap decision tree.
*   Gradient Boosting: Algoritma ensemble yang menghasilkan model prediktif dengan menggabungkan beberapa model prediktif yang lemah secara berurutan untuk memperbaiki kesalahan prediksi sebelumnya.
*   Logistic Regression: Menggunakan fungsi logistik (sigmoid) untuk memodelkan probabilitas bahwa sebuah instance termasuk dalam kelas tertentu.
*   Gaussian Naive Bayes:  Algoritma Naive Bayes yang mengasumsikan bahwa fitur-fitur mengikuti distribusi normal (Gaussian) dan menghitung probabilitas kelas berdasarkan nilai fitur-fitur yang diamati.

Metrik yang akan digunakan untuk evaluasi model tugas klasifikasi adalah accuracy, precision, recall, f1-score, dan auc-roc

```python
column_drop=[
    'veil-type', 'cap-shape', 'gill-attachment'
]
df = df.drop(column_drop, axis = 1)
```

Menentukan Variabel Dependen (y) dan Independen (X)

```python
X = df.drop('class', axis = 1)
y = df['class']
```

Resampling Data

```python
sm = SMOTE(sampling_strategy = 'auto', random_state = 3)
x, y = sm.fit_resample(X, y)
unique_values, counts = np.unique(y, return_counts=True)
for value, count in zip(unique_values, counts):
    print(f"Nilai: {value}, Frekuensi: {count}")
```

Split Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 3, stratify = y)
```

```python
#menampilkan baris dan kolom masing-masing data split
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Normalisasi

```python
sc = StandardScaler()

x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
```

# Model Training | Pelatihan Model

Melatih model dengan 5 algoritma untuk tugas klasifikasi, diantaranya decision tree, random forest, gradient boosting, logistic regression, dan naive bayes

```python
clf_tree = DecisionTreeClassifier()
clf_tree.fit(x_train_sc, y_train)
clf_tree.score(x_test_sc, y_test)
```

```python
clf_forest = RandomForestClassifier(n_estimators = 2)
clf_forest.fit(x_train_sc, y_train)
clf_forest.score(x_test_sc, y_test)
```

```python
clf_gd = GradientBoostingClassifier(n_estimators = 2)
clf_gd.fit(x_train_sc, y_train)
clf_gd.score(x_test_sc, y_test)
```

```python
clf_lr = LogisticRegression()
clf_lr.fit(x_train_sc, y_train)
clf_lr.score(x_test_sc, y_test)
```

```python
clf_nb = GaussianNB()
clf_nb.fit(x_train_sc, y_train)
clf_nb.score(x_test_sc, y_test)
```

Selisih akurasi masing-masing algoritma model training sedikit sekali (di bawah 10%). Maka dari itu, dipilih algoritma decision tree dan random forest dengan akurasi terbesar, yaitu 100% untuk menentukan hyperparameter turing

```python
cvs = StratifiedShuffleSplit(n_splits = 4, test_size = 0.3, random_state = 3)

#cross validation decision tree
score = cross_val_score(clf_tree, x, y, cv = cvs)
score
```

```python
#cross validation random forest
score2 = cross_val_score(clf_forest, x, y, cv = cvs)
score2
```

### Random Forest

### Melatih model dengan parameter terbaik

```python
models = {
    "Decision Tree": DecisionTreeClassifier(max_depth=7, random_state=3),
    "Random Forest": RandomForestClassifier(max_depth=9, n_estimators=3, random_state=3)
}

for name, model in models.items():
    model.fit(x_train_sc, y_train)
    score = model.score(x_test_sc, y_test)
    print(f"Model {name} Score:", score)
```

# Model Evaluation | Evaluasi Model

Menggunakan metrik accuracy, precision, recall, f1-score, dan auc-roc

```python
for name, model in models.items():
    report = classification_report(y_test, model.predict(x_test_sc))
    print(f"===Classification Report===")
    print(f"{name}: ")
    print(report)
```

```python
from sklearn.metrics import roc_curve, auc

for name, model in models.items():
    fpr, tpr, thresholds = roc_curve(y_test, model.predict(x_test_sc))
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label='%s (area = %0.2f)' % (name, roc_auc))

# Plot kurva ROC
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")
plt.show()
```

Fitur importance algoritma DT dan RF

```python
def calculate_fi(model, x_train):
    # Menghitung feature importance
    feature_coefficients = model.feature_importances_

    # Membuat DataFrame feature importance
    feature_importance_df = pd.DataFrame({
        'Feature': x_train.columns,
        'Coefficient': feature_coefficients
    })

    # Menghitung nilai absolut koefisien
    feature_importance_df['Absolute_Coefficient'] = feature_importance_df['Coefficient'].abs()

    # Mengurutkan DataFrame berdasarkan nilai absolut koefisien
    feature_importance_df = feature_importance_df.sort_values(by='Absolute_Coefficient', ascending=False)

    return feature_importance_df
```

```python
#fungsi untuk menampilkan plot feature importance
def plot_feature_importance(feature_importance_df, title):
    sns.set_style('dark')
    custom_palette = sns.color_palette("husl", len(feature_importance_df))
    sns.barplot(data=feature_importance_df, x='Absolute_Coefficient',
                y='Feature', hue='Feature', palette=custom_palette, legend=False)
    plt.title(title)
    plt.xlabel("Feature Importance")
    plt.show()
```

```python
# Menghitung feature importance dan plot untuk model Decision Tree dan Random Forest
for name, model in models.items():
    feature_importance_df = calculate_fi(model, x_train)
    plot_feature_importance(feature_importance_df, f"Feature Importance for {name}")
```
